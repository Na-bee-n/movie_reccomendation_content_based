```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
# Merging the two datasets on 'title'
movies = movies.merge(credits,on='title')
```

```python
movies.shape
```

```python
# FEATURE SELECTION:- Keeping onlt the relevant columns
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.shape
```

```python
movies.head()
```

```python
# Checking the null
movies.isnull().sum()
```

* In overview columns, there are three null values. This columns contains the description about movie which is important to consider. So it is better to drop the null values.

```python
# dropping the null values
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
import ast  # Abstract Syntax Trees: used to evaluate the python literals like strings,lists
```

```python
# function to convert genres column into list of Genres only from dict type
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
# function that return the list of top 3 cast only
def convert3(text):
    L = []
    counter =0
    for i in ast.literal_eval(text):
        if counter < 3 :
            L.append(i['name'])
        counter+=1
    return L
        
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
movies.head()
```

```python
def fetch_director(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.sample(2)
```

```python
# function that removes white space of an item inside list like ['Nabin Shrestha'] -->['NabinShrestha']. This treat as an individual entity 
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres']=movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
# a new column named tags is added storing all the columns into sigle one
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.sample(1)
```

```python
# dropping the columsn that are added into tag columns
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new.head()
```

```python
new['tags']=new['tags'].apply(lambda x:" ".join(x))
new.sample(2)
```

```python
# Using tfidf for vectorization of the data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python

def recommend(movie):
    index = new[new['title'] ==movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
